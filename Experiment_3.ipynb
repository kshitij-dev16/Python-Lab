```python
# Access Items
Championship =["Ferrari","Mercedes","Red Bull"]
print(Championship[:2])
```

```python
# Reverse/Negative Indexing
Championship =["Ferrari","Mercedes","Red Bull","Williams","Cadillac","Audi"]
print(Championship[-3])
```

```python
# Slicing
Championship =["Ferrari","Mercedes","Red Bull","Williams","Cadillac","Audi"]
print(Championship[-3:])
```

```python
# Change Item Value
Championship =["Ferrari","Mercedes","Red Bull","Williams","Cadillac","Audi"]
Championship[1] = "Haas"
Championship
```

```python
# Append Items
Championship =["Ferrari","Mercedes","Red Bull","Williams","Cadillac","Audi"]
Championship.append("McLaren")
Championship
```

```python
# Insert Items
Championship =["Ferrari","Mercedes","Red Bull","Williams","Cadillac","Audi"]
Championship.insert(2,"McLaren")
Championship
```

```python
# Extend List
Championship =["Ferrari","Mercedes","Red Bull","Williams","Cadillac","Audi"]
cars = ["McLaren","Haas"]
Championship.extend(cars)
Championship
```

```python
numbers = [10,20,30,40,50]
# Slicing examples
"Elements from index 1 to 3:"
print(numbers[1:3])
"First three elements:"
print(numbers[:4])
"Elements from index 2 onwards:(Reverse Indexing):"
print(numbers[-3:])
"All elements:"
print(numbers[:])
```

```python
marks = [78,85,90,66,88]
print("Maximum marks is: ",max(marks))
print("Minimum marks is: ",min(marks))
print("Sum of marks is: ",sum(marks))
print("Length of marks is: ",len(marks))
print("Average marks is: ",sum(marks)/len(marks))
marks.sort()
print("sorted marks in ascending order are:", marks)
```

```python
'''2. Grocery shopping List
A user maintains a grocery list for monthly shopping.'''
# Tasks:
# ## Create a grocery list
grocery_list = ["Milk","Sugar","Fruits","Eggs"]
print("grocery list is",grocery_list)
### Add new items to the list
grocery_list.append("Cheese")
print("grocery list with new items is",grocery_list)
### Remove purchsed items
grocery_list.remove("Milk")
print("grocery list after purchasing is",grocery_list)
### Display the final shopping list
print("final shopping list is",grocery_list)
```
