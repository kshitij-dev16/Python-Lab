# Sets

```python
# Creation of Set
thisset = {"Messi", "Suarez", "Neymar"}
print(thisset)
```

```python
# Duplicates not allowed
thisset = {"Messi", "Suarez", "Neymar", "Messi"}
print(thisset)
```

```python
# The values True and 1 are considered the same value in sets, and are treated as duplicates:
thisset = {"Messi", "Suarez", "Neymar", True, 1, 2}
print(thisset)
```

```python
# A Set with strings,integers and boolean values:
set1 = {"xyz", 14, False, 40, "Female"}
print(set1)
```

```python
# To check memebrship
thisset = {"Messi", "Suarez", "Neymar"}
print("Messi" in thisset)
print("Ronaldo" in thisset)
```

```python
# Once a set is created, you cannot replace its items,
# but you can add new items.
thisset = {"Messi", "Suarez", "Neymar"}
thisset.add("Iniesta")
print(thisset)
```

```python
# Remove Item
thisset = {"Messi", "Suarez", "Neymar"}
thisset.remove("Suarez")
thisset
```

```python
# Set Operations
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}
print("Union:", A | B)
print("Intersection:", A & B)
print("Difference:", B -A)
print("Symmetric Difference:", A ^ B)
```

```python
''' Problem Statement 2 : Three students choose elective subjects.
Find the common subjects chosen by all students using set operations.
concept used : Set intersection (&)'''

student1 = {"Physics","Mathematics","Biology","Psychology","Hisotry"}
student2 = {"Physics","Mathematics","Chemistry","Geography","Economics"}
student3 = {"Physics","History","Chemistry","Geography","Economics"}
Intersection = student1 & student2 & student3

print ("The common subject is", Intersection, student1&student2&student3)
```

```python
''' Problem Statement 3: One set contains students in the cricket club,another in the football club.
Find students who are:1. In both clubs 2. In only one club
Concept Used: Intersection,Symmetric difference'''

Cricket_club = {"Khush","Harshit","Dhvan","Krishiv","Swastik"}
Football_club = {"Khush","Swastik","Kshitij","Medhansh","Vishvam"}
Intersection = Cricket_club & Football_club
print("Students in both clubs are",Intersection)
Symmetric_difference = Cricket_club ^ Football_club
print("Students in only one club are",Symmetric_difference)
```

```python
''' Problem statement 4 : Absent Students
Given as set of all students and a set of presnet students, find the absent students.
Concept Used: Difference(-)'''
Original_students = {"Khush","Harshit","Dhvan","Krishiv","Swastik","Kshitij","Medhansh","Vishvam"}
Present_students = {"Khush","Harshit","Dhvan","Krishiv","Swastik"}
Absent = Original_students - Present_students
print("Absent students are",Absent)
```
